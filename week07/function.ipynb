## 함수의 변수 범위

* 함수 내에서 정의 된 변수는 함수 내에서만 존재

```python
var = None
del var
```

```python
def my_func():
    # 함수 내에서만 존재함
    var = 10
    print(var) # 얘 때문에 10 출력
    
my_func() # 10

print(var) # error. var가 정의 안됨
```

* 지역 변수를 먼저 봄
* 전역 변수를 봄
* 함수 내에서 전역 변수는 값을 가져다 쓸 수는 있지만, 값을 변화시킬 수는 없음

```python
var = 10 # 전역변수
print("1. 함수 밖 전역변수의 값",var) # 10

def my_func(var):
    # 전역변수와 이름은 같지만 my_func 안에서만 동작하는 지역변수임
    print('1. 함수 내 지역변수',var) # 10
    var += 10 # 20
    print('2. 함수 내 지역변수',var) # 20

my_func(var)

print("2. 함수 밖 전역변수의 값",var) # 10
```

* 전역 변수와 지역 변수는 이름을 달리해서 구분
* 함수 내의 변수 이름 앞에 언더라인(_)을 붙여주는 방법이 있음

```python
def my_func(_var):
    print('1. 함수 내 지역변수',_var) # 전역변수와 이름은 같지만 my_func 안에서만 동작하는 지역변수임
    _var += 10
    print('2. 함수 내 지역변수',_var)

var = 10
print("1. 함수 밖 전역변수의 값",var)
my_func(var)

print("2. 함수 밖 전역변수의 값",var)
```

```python
def my_func():
    # 전역 변수를 함수 내에서 변화시키려면 global 을 사용
    global var
    print("1. 함수 내에서 값",var)
    var += 10
    print("2. 함수 내에서 값",var)
    
var = 10
print("1. 함수 밖에서 값",var)
my_func()
print("2. 함수 밖에서 값",var)
```

## 유연한 함수 정의 방법: 인자 개수 정하지 않고 함수 정의

1. 기본 인자 값: 하나 이상의 인자들에 기본값을 미리 정해주는 것

```python
def ask_ok(prompt, retries=4, reminder="다시 입력해주세요!"):

    while True:
        ok = input(prompt)

        if ok in ('y'):
            return True
        if ok in ('n'):
            return False
        
        retries -= 1
        
        if retries == 0:
            print("잘못된 입력")
            break
        print(reminder)
```

```python
# 반드시 필요한 인자만 전달
# prompt 에 전달됨
ask_ok("종료할까요?")
```

```python
# 선택적 인자 하나를 제공
# prompt, retries 에 전달
ask_ok("파일을 덮어쓸까요?", 2)
```

```python
# 모든 인자를 제공
ask_ok("파일을 덮어쓸까요?",2,"y 또는 n로 답하세요!")
```

* 기본 인자 값은 함수가 정의될 때만 구해짐

```python
i = 5

# i = 5 인 타이밍에 함수가 정의되었음
def my_func(arg = i): 
    print(arg)
    
i = 6 # i가 값이 바뀌었지만

my_func() # 화면에는 5가 출력 됨
```

* 기본 인자가 함수가 정의될 때 구해지므로, 리스트나 딕셔너리와 같은 가변 객체 일 때 아래와 같은 일이 일어남

```python
# L 이라는 리스트는 함수가 정의될 때 한 번 구해짐
def my_func(a, L = list()):
    L.append(a)
    print(L)

# 함수가 호출 될 때, L 리스트가 새롭게 만들어지지 않는다는 의미
# 따라서 L 리스트에는 함수가 호출될 때 값이 누적 됨
my_func(1)
my_func(2)
my_func(3)
```

```python
# a 는 아무일도 하지 않음
def my_func(a, L=[1,2,3]):
    # pop은 리스트의 맨 마지막 요소를 뱉고, 삭제
    print(L.pop()) 
    
my_func(1)
my_func(1)
my_func(1)
# L이 비어있는데, pop 시키기 때문에 에러
my_func(1) 
```

* 기본값이 공유되지 않도록 하려면, 기본 인자 값에 None을 줌

```python
def my_func(a, L = None):
    if L is None:
        L = []
    L.append(a)
    print(L)

my_func(1)
my_func(2)
my_func(3)

# 기본 인자 None 대신에 리스트 [1,2] 가 입력됨
my_func(3, [1,2]) 
```

3. 임의의 인자 목록: 사용 빈도 낮지만 코드 읽기를 위해 알아둘 것

```python
# 세 개의 인자를 받아 총 합을 출력하는 함수
def adder(x,y,z):
    print("sum:",x+y+z)
    
adder(3,4,5)

# 하지만 셋을 초과하는 인자를 주면 에러 발생
adder(3,4,5,6)
```

```python
# 이 때 *args 를 사용
# 변수 이름 앞에 별표(*)를 붙임
# 여러 값들이 튜플로 묶여서 입력 됨

def adder(*num): # num 은 튜플
    summation = 0
    
    for n in num:
        summation += n
    
    print(summation)
    
# 인자의 개수 상관없이 사용 가능
adder(1)
adder(1,2,3)
adder(1,2,3,4,5,6,7)
```

* \*args 쓰는 파이썬의 내장 함수는 print() 가 있음

* print(*objects, sep=' ', end='\n', file=sys.stdout, flush=False)
 * 의미: objects 를 화면에 나타내고, 여러 인자는 sep로 구분, 마지막에는 end 를 붙임
 * 모든 인자는 str()로 캐스팅과 같이 문자열로 변환 됨
 * 기본 인자는 sep 은 공백 구분, end 는 줄바꿈

```python
print(1)
print(1,2,3)
print(1,2,3,sep='') # 문자열 구분 공백 없이 출력
print(1,2,3,end='END') # 마지막에 줄바꿈 대신에 END를 붙임
```

```python
# 정해지지 않은 개수의 키, 값 쌍을 입력할 수 있음
# **kwargs 를 사용
# 변수 앞에 별표(*)를 둘 붙임 (**)
# 여러 값들이 딕셔너리로 묶여서 입력 됨

def intro(**data):
    
    for k, v in data.items():
        print(f'{k} 는 {v} 입니다.')
        
intro(Firstname='Sita', Lastname='Sharma', age=22, phone=1234567890)
```

## 람다 표현식
* lambda 키워드로 이름 붙일 필요 없이 작은 함수를 만들 수 있음
* lambda 를 쓰고, 콜론(:)으로 구분하여
* 콜론 왼쪽에는 함수에 입력 될 인자
* 콜론 오른쪽에는 함수에서 반환해야할 인자 씀

```python
# f라는 람다 함수 정의
f = lambda x : x + 10
print(f(5))
```

```python
# 이름을 따로 만들지 않고 바로 사용 가능
(lambda x: x+10)(10)
```

```python
# 람다 표현식 안에 새로운 변수를 만들 수 없음
# 외부 변수 사용은 가능
y = 10
f = lambda x : x + y
print(f(5))
```

```python
# 함수에서 반환 값에 사용 가능
def make_incrementor(n):
    return lambda x:x + n

f = make_incrementor(5) # 입력 값에 5를 더해주는 함수를 정의
print(f(5))
print(f(10))
```

## 클로져
* 파이썬에서는 함수 안에 함수를 정의할 수 있음
* 내용을 각자 읽어보시고, 한 줄씩 익히기 바랍니다.

```python
# 바깥 쪽 함수
def print_hello():
    message = "Hello!"
    
    # 안쪽 함수
    def print_message():
        print(message)
    
    # 바깥 쪽 함수에서 안쪽 함수를 실행
    print_message()
    
print_hello() # "Hello!"
```

```python
# 바깥쪽 함수
def outer():
    x = 10 # 바깥 함수에서 변수 선언
    
    def inner():
        x = 20 # 바깥 함수의 변수와 이름이 같지만, 안쪽 함수에서만 동작함
        
    inner() # inner함수를 실행
    print(x) # inner함수의 x는 바깥쪽 함수의 변수에 영향 안줌. 따라서 outer x = 10
    
outer()
```

```python
def outer():
    a = 10 # 바깥쪽 함수에서 변수 선언
    
    def inner():
        b = 20
        print(a+b) # outer a도 볼 수 있으므로 30

    inner() # inner함수 실행하면 화면에 30 출력

outer()
```

```python
def outer():
    a = 10
    
    def inner():
        b = 20
        a += b # outer a는 inner 밖에 있어서 변경 불가
        print(a)
        
    inner()
    
outer()
```

```python
def outer():
    a = 10
    
    def inner():
        nonlocal a # inner 밖의 a를 변경하려면 nonlocal 붙여줌
        b = 20
        a += b
        print(a)
        
    inner()
    
outer()
```

1.
 * 피보나치 수열을 N 미만의 원소 까지 출력하는 함수 작성

```python
def fib(n):
    a, b = 0, 1
    
    while a < n:
        print(a, end=' ')
        a, b = b, a+b
        
fib(1000)
```

2.
 * 아래와 같은 기능을 하는 함수 만들기
 * 입력인자: 정수를 입력 받음
 * 함수가 할 일: 주사위를 20번 굴려서 리스트에 넣음
 * 반환값: 주사위 20회 굴리기 중에서 입력 된 인자와 일치하는 횟수

```python
import random

def dice(num):
    _dice = [random.randrange(1,7) for _ in range(20)]
    print(sorted(_dice))
    return _dice.count(num)

print(dice(3))
```
